```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from parse_into_pandas import DictDF
```

```python
ddf = DictDF()
```

```python
ddf.load_single_entries()
```

```python
#ddf.s_df.head()
#ddf.s_df.columns
```

```python
group_tones = ddf.s_df.groupby('tone')
#group_tones.size()
```

```python
group_pinraw = ddf.s_df.groupby('pinyin(raw)')
```

```python
group_pinyin = ddf.s_df.groupby('pinyin')
#group_pinyin.size()
```

```python
max_tone = (group_tones.size().idxmax(), group_tones.size().max())
max_tone
```

```python
max_pinraw = (group_pinraw.size().idxmax(), group_pinraw.size().max())
max_pinraw
```

```python
max_pinyin = (group_pinyin.size().idxmax(), group_pinyin.size().max())
max_pinyin
```

```python
plt.bar(range(1,len(group_tones.size())+1),group_tones.size())
plt.ylabel('Frequency')
plt.title('Tones frequency')

plt.show()
```

```python
pinraw = group_pinraw.size().sort_values(ascending=False)
p1 = plt.bar(range(len(pinraw)), pinraw)
p2 = plt.plot(np.arange(len(pinraw)),[pinraw.median()]*len(pinraw)) 
p3 = plt.plot(np.arange(len(pinraw)),[pinraw.quantile(q=0.75)]*len(pinraw), '--') 
p4 = plt.plot(np.arange(len(pinraw)),[pinraw.quantile(q=0.25)]*len(pinraw), '--') 

plt.ylabel('Frequency')
plt.title('Raw pinyin frequency (median = {})'.format(pinraw.median()))
#plt.legend(p1[0], 'Median value: {}'.format(pinraw.median()))

plt.show()
```

```python
pinyin = group_pinyin.size().sort_values(ascending=False)

p1 = plt.bar(range(len(pinyin)), pinyin)
p2 = plt.plot(np.arange(len(pinyin)),[pinyin.median()]*len(pinyin)) 
p3 = plt.plot(np.arange(len(pinyin)),[pinyin.quantile(q=0.75)]*len(pinyin), '--') 
p4 = plt.plot(np.arange(len(pinyin)),[pinyin.quantile(q=0.25)]*len(pinyin), '--') 

plt.ylabel('Frequency')
plt.title('Pinyin frequency (median = {})'.format(pinraw.median()))

plt.show()
```
